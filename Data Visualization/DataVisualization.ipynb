**Box plot**: Box is representation of data in form of diagram where data values are represented as Box and we can infer the maximun, minium and vlaues that lies between some values.

**Histogram**: Graphical representation of distribution of the numerical data.

## II.1. Introduction of data for Classification

Data set for Classification used is related to the direct marketing campaigns of a Portuguese banking institution. The marketing campaigns were based on phone calls. Often more than one phone call is required to the same customer to access if the product (Bank term deposit) would be subscribed or not Below is the information about the dataset: -
          
**Dataset name**: bank-full.csv

**Number of features**: 17

**Number of Instances**: 45211
         
**Information about features**: -

Input variables:
    bank client data:
1. age (numeric)
2. job : type of job (categorical: "admin.","unknown","unemployed","management","housemaid","entrepreneur","student", "blue-collar","self-employed","retired","technician","services")                       
3. marital : marital status (categorical: "married","divorced","single"; note: "divorced" means divorced or widowed)
4. education (categorical: "unknown","secondary","primary","tertiary")
5. default: has credit in default? (binary: "yes","no")
6. balance: average yearly balance, in euros (numeric) 
7. housing: has housing loan? (binary: "yes","no")
8. loan: has personal loan? (binary: "yes","no")
    related with the last contact of the current campaign:
9. contact: contact communication type (categorical: "unknown","telephone","cellular") 
10. day: last contact day of the month (numeric)
11. month: last contact month of year (categorical: "jan", "feb", "mar", ..., "nov", "dec")
12. duration: last contact duration, in seconds (numeric)
   
   other attributes:
13. campaign: number of contacts performed during this campaign and for this client (numeric, includes last contact)
14. pdays: number of days that passed by after the client was last contacted from a previous campaign (numeric, -1 means client was not previously contacted)
15. previous: number of contacts performed before this campaign and for this client (numeric)
16. poutcome: outcome of the previous marketing campaign (categorical: "unknown","other","failure","success")

  Output variable (desired target):
17. y - has the client subscribed a term deposit? (binary: "yes","no")

```python
# Coverting object data type to numeric 
jobtype = {'management': 5, 'technician': 1, 'entrepreneur': 2, 'blue-collar': 3,
       'unknown': 0, 'retired': 6, 'admin.': 7, 'services': 8, 'self-employed': 9,
       'unemployed': 10, 'housemaid': 11, 'student': 12}

maritaltype = {'married': 0, 'single': 1, 'divorced': 2}

educationtype = {'tertiary': 2, 'secondary': 1, 'unknown': 0, 'primary': 3}

ytype = {'no': 0, 'yes': 1}

contacttype = {'unknown': 0, 'cellular': 1, 'telephone': 2}

monthtype = {'may': 5, 'jun': 6, 'jul': 7, 'aug': 8, 'oct': 10, 'nov': 11, 'dec': 12, 'jan': 1, 'feb': 2,
       'mar': 3, 'apr': 4, 'sep': 9}

poutcometype = {'unknown': 0, 'failure': 1, 'other': 2, 'success': 3}

df_Class['job'] = df_Class['job'].apply(lambda x: jobtype[x])
df_Class['marital'] = df_Class['marital'].apply(lambda x: maritaltype[x])
df_Class['education'] = df_Class['education'].apply(lambda x: educationtype[x])
df_Class['default'] = df_Class['default'].apply(lambda x: ytype[x])
df_Class['housing'] = df_Class['housing'].apply(lambda x: ytype[x])
df_Class['loan'] = df_Class['loan'].apply(lambda x: ytype[x])
df_Class['contact'] = df_Class['contact'].apply(lambda x: contacttype[x])
df_Class['month'] = df_Class['month'].apply(lambda x: monthtype[x])
df_Class['poutcome'] = df_Class['poutcome'].apply(lambda x: poutcometype[x])
df_Class['y'] = df_Class['y'].apply(lambda x: ytype[x])

df_Class.head()
```
